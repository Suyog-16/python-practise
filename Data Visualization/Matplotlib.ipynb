## Visulaization using Matplotlib
 we use the most popular module pyplot here.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Reading data from a csv file

```python
df = pd.read_csv('house_price.csv')
df.head()
```

There are many types of plots you can make in matplotlib
Some of them are:
1) Scatter Plot
2) Bar Plot
3) Histogram Plot
4) Box plot
5) Line plot

## Scatter plot

```python
plt.scatter(df['Price'],df['Size'])
plt.title(" Price vs Size")
plt.xlabel("Price (in $1000)")
plt.ylabel("Size(Sq ft)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,9))
plt.hist(df['Size'],edgecolor='black',color='lightblue')
plt.title("Distribution of sizes")
plt.xlabel("House Sizes(sq ft)")
plt.ylabel("House Sizes(sq ft)")
plt.grid(True)
```
